```python
#gets mps voting patterns
url = 'http://www.publicwhip.org.uk/data'
req = request.Request(url, headers={'User-Agent':'Mozilla/5.0'})
response = request.urlopen(req)
page = response.read()
soup = bs(page)
```

```python
links = soup.findAll('a')
for item in links[:]:
    if 'votematrix' not in str(item) or 'lords' in str(item):
        links.remove(item)

       
```

```python
votes = {}
info = {}
data_values = {}
for item in links[:2]:
    name = item['href'][-8:-4]
    if 'dat' in str(item):
        req = request.Request(url+'/'+item['href'], headers={'User-Agent':'Mozilla/5.0'})
        response = request.urlopen(req)
        csvstr = str(response.read()).strip("b'")
        plines = csvstr.split('\\t')
        columns = []
        data = []
        for item in plines:
            if item[:4] in ['rowi','date','vote','Bill','mpid']:
                columns.append(item.strip('\\n').strip(' \\n'))
            else:
                data.append(item.strip('\\n').strip(' \\n'))
                
        i = len(columns)
        j = 1
        start = 0
        stacked_data = []
        while True:
            i =  len(columns)*j
            if i > len(data):
                break
            else:
                stacked_data.append(data[start:i])
                start = i
                j +=1
        votes.update({name:DataFrame(stacked_data, columns = columns)})
    else:
        req = request.Request(url+'/'+item['href'], headers={'User-Agent':'Mozilla/5.0'})
        response = request.urlopen(req)
        csvstr = str(response.read()).strip("b'").replace('\\n','\\t')
        plines = csvstr.split('\\t')
         
        columns = []
        data = []
        for i in range(len(plines[:100])):
            if 'mpid' == str(plines[i]):
                mp_start = i
                print(mp_start)
        if len(data_values) == 0:
            data_values.update({'vote_codes':plines[:mp_start]})
        for item in plines[mp_start:]:
            if item[:4] in ['mpid','firs','surn','part','Publ']:
                columns.append(item.strip('\\n').strip(' \\n'))
            else:
                data.append(item.strip('\\n').strip(' \\n'))
            
        i = len(columns)
        j = 1
        start = 0
        stacked_data = []
        while True:
            i =  len(columns)*j
            if i > len(data):
                break
            else:
                stacked_data.append(data[start:i])
                start = i
                j +=1
        info.update({name:DataFrame(stacked_data, columns = columns)})
        
            
                
        
            
        
```

```python
#gets MPs debates
API_KEY = 'BhSfveCajnJbCE2p83At7Eo6'
API_URL = 'http://www.theyworkforyou.com/api/'
API_OUTPUT = 'js'

def make_api(endpoint, params):
    '''constructs an api url from inputs'''
    url = "{}{}".format(API_URL, endpoint)
    params['key'] = API_KEY
    if 'output' not in params:
        params['output'] = API_OUTPUT
    response = rq.get(url, params)
    return response
```
