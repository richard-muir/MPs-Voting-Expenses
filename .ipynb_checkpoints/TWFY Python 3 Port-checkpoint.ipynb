```python
results = {}
for mp in mp_list:
    party = mp['party']
    if party in results.keys():
        results[party] += 1
    else:
        results[party] = 1
```

```python
other = 0
parties = set()
for k, v in results.items():
    print(k,v)
    if v < 10:
        parties.add(k)
        print(v)
        other += v

        
for party in parties:
    del results[party]
    
results['Other'] = other
```
