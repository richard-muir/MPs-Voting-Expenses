```python
#gets MPs debates
API_KEY = 'BhSfveCajnJbCE2p83At7Eo6'
API_URL = 'http://www.theyworkforyou.com/api/'
API_OUTPUT = 'js'

def make_api(endpoint, params):
    '''constructs an api url from inputs'''
    url = "{}{}".format(API_URL, endpoint)
    params['key'] = API_KEY
    if 'output' not in params:
        params['output'] = API_OUTPUT
    response = rq.get(url, params)
    return response
```
